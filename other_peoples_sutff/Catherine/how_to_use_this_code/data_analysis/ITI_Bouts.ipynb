```python
# Load the csvs - sniffing data
behavior22_path= r"C:\Users\zhaoz\Desktop\Research\data\aggregated_csvs\22_object_CTBN.csv"
behavior22 = pd.read_csv(behavior22_path)

# Get the timestamps of mouse in and mouse out
mouse_in = np.array(behavior22[behavior22['Behavior'] == 'mice back in']['Start (s)'].tolist())
mouse_out = np.array(behavior22[behavior22['Behavior'] == 'mice taken out']['Start (s)'].tolist())

# Convert the dataframes into arrays of behaviors for exp4
exp4_df = behavior22[(behavior22['Start (s)'] > mouse_in[3])]
exp4_facial_sniffing_bouts = boris.get_behavior_bouts(exp4_df,
										  subject=['subject'],
										  behavior=['facial sniffing'], 
										  min_iti = 0,
										  min_bout= 0)
exp4_anogenital_sniffing_bouts = boris.get_behavior_bouts(exp4_df,
										  subject=['subject'],
										  behavior=['anogenital sniffing'], 
										  min_iti = 0,
										  min_bout= 0)

# Manipulate the min ini and about for exp4
exp4_facial_thresholded = bet.threshold_bouts(exp4_facial_sniffing_bouts, min_iti = 1, min_bout = .5)
exp4_anogenital_thresholded = bet.threshold_bouts(exp4_anogenital_sniffing_bouts, min_iti = 1, min_bout = .5)
bet.plot_event_bars(eventA = exp4_facial_sniffing_bouts[:len(exp4_anogenital_sniffing_bouts)], 
					event_dict = {'exp4 facial thresholded': exp4_facial_thresholded[:len(exp4_facial_thresholded)], 'exp4 anogenital thresholded': exp4_anogenital_thresholded[:len(exp4_anogenital_thresholded)]},
					eventB = exp4_anogenital_sniffing_bouts[:len(exp4_anogenital_sniffing_bouts)],
					title = '22 ITI and small bout manipulation')
```
