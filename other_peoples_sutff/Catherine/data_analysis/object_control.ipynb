```python
def plot_sniffing(data, ylabel, title):
	x = np.arange(len(conditions))
	width = 0.25
	multiplier = 0

	fig, ax = plt.subplots(layout = 'constrained')

	for behavior, measurement in data.items():
		offset = width*multiplier
		rects = ax.bar(x + offset, measurement['mean'], width, yerr=measurement['sem'], label = behavior)
		ax.bar_label(rects, padding = 3)
		multiplier += 1

	plt.ylim(0, max(v for group in data.values() for tup in group.values() for v in tup) * 1.5)
	ax.set_ylabel(ylabel)
	ax.set_title(title)
	ax.set_xticks(x + width, conditions)
	ax.legend(loc='best')
	plt.show()
```
