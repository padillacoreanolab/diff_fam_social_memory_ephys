```python
import numpy as np

def find_event_overlaps(event1, event2, percent_overlap=0.5, return_nonoverlap=False, mode='overlap'):
    """
    Find overlapping periods between two arrays of start and stop times.

    Parameters:
    -----------
    event1 : numpy.ndarray
        Array of start and stop times for event1, shape (n, 2)
    event2 : numpy.ndarray
        Array of start and stop times for event2, shape (m, 2)
    percent_overlap : float, optional
        Minimum percentage of overlap required, between 0 and 1. Default is 0.5 (50% overlap).
    return_nonoverlap : bool, optional
        If True, returns additional arrays for non-overlapping periods.
        If False, only returns overlapping events. Default is False.
    mode : str, optional
        Determines how overlapping events are processed:
        - 'combination': Creates events using earliest start and latest stop times
        - 'split': Creates events of only the overlapping portions
        - 'overlap': Same as 'split', returns only overlapping parts
        - 'duplicate': Returns original events sorted by whether they overlap
        Default is 'overlap'.

    Returns:
    --------
    Depending on parameters, returns various combinations of these arrays:
    event3 : numpy.ndarray - Overlapping events as determined by mode
    event4 : numpy.ndarray - Event1 parts that don't overlap with event2 (if return_nonoverlap=True)
    event5 : numpy.ndarray - Event2 parts that don't overlap with event1 (if return_nonoverlap=True)
    event6 : numpy.ndarray - Original event1 intervals that overlap (if mode='duplicate')
    event7 : numpy.ndarray - Original event2 intervals that overlap (if mode='duplicate')
    """
    # Check for empty input arrays
    if len(event1) == 0 or len(event2) == 0:
        if mode == 'duplicate':
            if return_nonoverlap:
                return np.empty((0, 2)), np.empty((0, 2)), np.empty((0, 2)), event1.copy(), event2.copy()
            else:
                return np.empty((0, 2)), np.empty((0, 2)), np.empty((0, 2))
        elif return_nonoverlap:
            return np.empty((0, 2)), event1.copy(), event2.copy()
        else:
            return np.empty((0, 2))

    # Initialize lists to store results
    combined_overlaps = []     # For 'combination' mode
    exact_overlaps = []        # For 'split' and 'overlap' modes
    event1_overlaps = []       # For 'duplicate' mode
    event2_overlaps = []       # For 'duplicate' mode

    # Make copies of events to track what remains non-overlapping
    if return_nonoverlap:
        # Convert to list of lists for easier manipulation
        event1_remaining = [list(e) for e in event1]
        event2_remaining = [list(e) for e in event2]

    # Check each pair of events for overlap
    for i, (start1, stop1) in enumerate(event1):
        duration1 = stop1 - start1

        for j, (start2, stop2) in enumerate(event2):
            duration2 = stop2 - start2

            # Calculate overlap
            overlap_start = max(start1, start2)
            overlap_stop = min(stop1, stop2)

            if overlap_start < overlap_stop:  # There is some overlap
                overlap_duration = overlap_stop - overlap_start

                # Calculate overlap ratios
                overlap_ratio1 = overlap_duration / duration1
                overlap_ratio2 = overlap_duration / duration2

                # Check if overlap meets threshold
                if overlap_ratio1 >= percent_overlap or overlap_ratio2 >= percent_overlap:
                    if mode == 'combination':
                        # Combined event using earliest start and latest stop
                        combined_overlaps.append([min(start1, start2), max(stop1, stop2)])

                        if return_nonoverlap:
                            # Remove this event from remaining lists
                            if [start1, stop1] in event1_remaining:
                                event1_remaining.remove([start1, stop1])
                            if [start2, stop2] in event2_remaining:
                                event2_remaining.remove([start2, stop2])

                    elif mode in ['split', 'overlap']:
                        # Exact overlapping period
                        exact_overlaps.append([overlap_start, overlap_stop])

                        if return_nonoverlap:
                            # Process event1 - find current event in remaining list
                            for k, e in enumerate(event1_remaining):
                                if e[0] == start1 and e[1] == stop1:
                                    # Replace with non-overlapping segments
                                    new_segments = []
                                    if start1 < overlap_start:
                                        new_segments.append([start1, overlap_start])
                                    if stop1 > overlap_stop:
                                        new_segments.append([overlap_stop, stop1])

                                    # Remove original segment and add new ones if any
                                    event1_remaining.pop(k)
                                    event1_remaining.extend(new_segments)
                                    break

                            # Process event2 - find current event in remaining list
                            for k, e in enumerate(event2_remaining):
                                if e[0] == start2 and e[1] == stop2:
                                    # Replace with non-overlapping segments
                                    new_segments = []
                                    if start2 < overlap_start:
                                        new_segments.append([start2, overlap_start])
                                    if stop2 > overlap_stop:
                                        new_segments.append([overlap_stop, stop2])

                                    # Remove original segment and add new ones if any
                                    event2_remaining.pop(k)
                                    event2_remaining.extend(new_segments)
                                    break

                    elif mode == 'duplicate':
                        # Exact overlapping period for event3
                        exact_overlaps.append([overlap_start, overlap_stop])

                        # Original events that overlap
                        event1_overlaps.append([start1, stop1])
                        event2_overlaps.append([start2, stop2])

                        if return_nonoverlap:
                            # Remove these events from remaining lists
                            if [start1, stop1] in event1_remaining:
                                event1_remaining.remove([start1, stop1])
                            if [start2, stop2] in event2_remaining:
                                event2_remaining.remove([start2, stop2])

    # Convert results to numpy arrays
    if mode == 'combination':
        event3 = np.array(combined_overlaps) if combined_overlaps else np.empty((0, 2))
    elif mode in ['split', 'overlap']:
        event3 = np.array(exact_overlaps) if exact_overlaps else np.empty((0, 2))
    elif mode == 'duplicate':
        event3 = np.array(exact_overlaps) if exact_overlaps else np.empty((0, 2))
        event6 = np.array(event1_overlaps) if event1_overlaps else np.empty((0, 2))
        event7 = np.array(event2_overlaps) if event2_overlaps else np.empty((0, 2))

    # Return results based on parameters
    if return_nonoverlap:
        event4 = np.array(event1_remaining) if event1_remaining else np.empty((0, 2))
        event5 = np.array(event2_remaining) if event2_remaining else np.empty((0, 2))

        if mode == 'duplicate':
            return event3, event6, event7, event4, event5
        else:
            return event3, event4, event5
    else:
        if mode == 'duplicate':
            return event3, event6, event7
        else:
            return event3
```
